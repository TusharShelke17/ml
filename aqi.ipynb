```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("city_day.csv")
data.head()
```

```python
data = data.fillna(data.mean())
```

```python
data.info()
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

# Group by date and calculate the mean AQI for each day
daily_mean_aqi = data.groupby('Date')['AQI'].mean()

# Plotting the overall AQI trend
plt.figure(figsize=(10, 6))
plt.plot(daily_mean_aqi.index, daily_mean_aqi.values, label='AQI')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('Overall AQI Trend over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(data["Date"], data["PM2.5"] , label= 'PM2.5')
plt.xlabel("PM2.5")
plt.ylabel("Date")
plt.title("pm2.5 vs date")
plt.legend()
plt.show()
```

```python
daily_meanaqi = data.groupby('Date')["AQI"].mean()

plt.figure(figsize=(10, 6))
plt.bar(daily_meanaqi.index, daily_meanaqi.values, label = "AQI")
plt.title("date vs aqi")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
plt.scatter(data['PM2.5'], data['AQI'], alpha=0.5)
plt.xlabel('PM2.5 Level')
plt.ylabel('AQI')
plt.title('Scatter Plot: PM2.5 vs. AQI')
plt.show()
```

```python
pollutants = [ "PM10",	"NO",	"CO",	"SO2", "O3"]

plt.figure(figsize=(5, 5))

for poll in pollutants:
  sns.scatterplot(x = poll, y = data["AQI"],data =data,  alpha = 0.5)
  plt.title(f'Scatter Plot: {poll} vs. AQI')
  plt.xlabel(f'{poll} Level')
  plt.ylabel('AQI')
  plt.show()
```
