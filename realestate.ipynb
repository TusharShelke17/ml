```python
subset1 = data[(data["transactiondate"]==2013) & (data["unit_area"]>25)]
subset1
```

```python
max_unit_area_by_stores = data.groupby('stores')['unit_area'].max()
max_unit_area_by_stores
```

```python
mean_unit_area_by_stores = data.groupby('stores')['unit_area'].mean()
mean_unit_area_by_stores
```

```python
Q1 = data['unit_area'].quantile(0.25)
Q3 = data['unit_area'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.3 * IQR
upper_bound = Q3 + 1.3 * IQR

# Identify outliers
outliers = data[(data['unit_area'] < lower_bound) | (data['unit_area'] > upper_bound)]

print(outliers)
```
